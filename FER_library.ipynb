```python
from fer import FER
import cv2
```

For image Facial expression recognition

```python
image_path = 'imagefolder/download (6).jfif'
image = cv2.imread(image_path)
detector = FER(mtcnn=True)

emotions = detector.detect_emotions(image)

for emotion, score in emotions[0]["emotions"].items():
    print(f"{emotion}: {score}")

bounding_box = emotions[0]["box"]

cv2.rectangle(image, (bounding_box[0], bounding_box[1]), (bounding_box[0] + bounding_box[2], bounding_box[1] + bounding_box[3]), (0, 0, 255), 2)

cv2.imshow("Image", image)
while True:
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break
cv2.waitKey(0)
cv2.destroyAllWindows()
```

for video facical recognition

```python
video_path = 'path_to_your_video_file.mp4'

detector = FER(mtcnn=True)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break


    emotions = detector.detect_emotions(frame)

    for emotion, score in emotions[0]["emotions"].items():
        print(f"{emotion}: {score}")

    bounding_box = emotions[0]["box"]

    cv2.rectangle(frame, (bounding_box[0], bounding_box[1]), (bounding_box[0] + bounding_box[2], bounding_box[1] + bounding_box[3]), (0, 0, 255), 2)

    cv2.imshow("Video Frame", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
